```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
image_path = "images/aur.jpg"
#original_image = Image.open(image_path)
inputpic = cv2.imread(image_path)
pic = cv2.cvtColor(inputpic, cv2.COLOR_BGR2RGB)
plt.imshow(pic)
plt.show()
plt.imshow(cv2.cvtColor(pic, cv2.COLOR_RGB2GRAY), cmap='gray')
plt.show()
```

```python
b = pic.copy()
# set green and red channels to 0
b[:, :, 1] = 0
b[:, :, 0] = 0

g = pic.copy()
# set blue and red channels to 0
g[:, :, 0] = 0
g[:, :, 2] = 0

r = pic.copy()
# set blue and green channels to 0
r[:, :, 1] = 0
r[:, :, 2] = 0

plt.imshow(b)
plt.show()

plt.imshow(r)
plt.show()

plt.imshow(g)
plt.show()
```
